## 1. 匯入套件

```python
import requests
import json
```

## 2.觀察網站


```python
url = 'https://www.costco.com.tw/rest/v2/taiwan/products/search?fields=FULL&query=%E6%B3%A1%E9%BA%B5&pageSize=9999&lang=zh_TW&curr=TWD'
data = requests.get(url).text
print(data)
```

```python
type(data)
```

## 3.解析


```python
json_data = json . loads(data)
```

## 4.輸出

```python
json_data . keys()
```

```python
json_data["products"][0]["price"]["formattedValue"]
```

```python
for i in range(10):
    print(json_data["products"][i]["name"])
    print(json_data["products"][i]["price"]["formattedValue"])
```
