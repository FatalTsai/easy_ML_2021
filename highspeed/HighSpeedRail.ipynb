```python
import requests, json
from bs4 import BeautifulSoup

payload = {
    'SearchType': 'S',
    'Lang': 'TW',
    'StartStation': 'NanGang',
    'EndStation': 'ZuoYing',
    'OutWardSearchDate': '2021/11/26',
    'OutWardSearchTime': '18:00',
    'ReturnSearchDate': '2021/11/27',
    'ReturnSearchTime': '19:00'
}

url = 'https://www.thsrc.com.tw/TimeTable/Search'
html = requests.post(url, data=payload).text
soups = BeautifulSoup(html, 'html.parser')

new_soups = json.loads(str(soups))
TrainNumber = []
DepartureTime = []
DestinationTime = []
Duration = []

total = len(new_soups['data']['DepartureTable']['TrainItem'])
print('車次', '出發時間', '抵達時間', '行車時間', sep = '\t')

for i in range(total):
    TrainNumber.append(new_soups['data']['DepartureTable']['TrainItem'][i]['TrainNumber'])
    DepartureTime.append(new_soups['data']['DepartureTable']['TrainItem'][i]['DepartureTime'])
    DestinationTime.append(new_soups['data']['DepartureTable']['TrainItem'][i]['DestinationTime'])
    Duration.append(new_soups['data']['DepartureTable']['TrainItem'][i]['Duration'])



    
```

```python
output = "車次,出發時間,抵達時間,行車時間\n"
for i in range(len(TrainNumber)):
    output += TrainNumber[i]+","+DepartureTime[i]+","+DepartureTime[i]+","+Duration[i]+"\n"


print(output)
```

```python
print(output)
with open("highspeed_big5.csv","w",encoding='Big5') as file:
    file.write(output)

with open("highspeed_utf-8.csv","w",encoding='utf-8') as file:
    file.write(output)
```
