```python
import requests 
import time 
from bs4 import BeautifulSoup
url= 'https://online.carrefour.com.tw/zh/search?q=餅乾'
html = requests.get(url).text 
soup = BeautifulSoup(html, 'html.parser')

output  = "name,price,time\n"

products = soup.find_all('div', 'desc-operation-wrapper')
# print(products)
for product in products: 
  # print('名稱', str(product.find('div','commodity-desc').a.text)+"," )
  output += str(product.find('div','commodity-desc').a.text)+","
  try: 
    # print('價錢', product.find('div','current-price').em.text+",")
    output += product.find('div','current-price').em.text+","
  except: 
    # print('未標價')
    output += "未標價,"

    

  output += str(time.ctime())+"\n"
  # print(output)

```

```python
print(output)
with open("carrefour_big5.csv","w",encoding='Big5') as file:
    file.write(output)

with open("carrefour_utf-8.csv","w",encoding='utf-8') as file:
    file.write(output)
```
